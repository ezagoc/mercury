```python
import pandas as pd
from tqdm import tqdm
import numpy as np

ba = 'SA'
df = pd.read_parquet(f'../../data/03-experiment/{ba}/baseline/01-preprocess/influencers/followers.parquet').rename(
columns = {'strong': 'strong_tie','weak': 'weak_tie'})
assing = pd.read_excel(f'../../data/04-analysis/{ba}/assignments_permutations.xlsx')

df['weak_tie'] = np.where(df['strong_tie']== 1, 0, df['weak_tie'])

```

```python
for e in range(0, 5):
    a = e*200 + 1
    b = (e+1)*200 + 1
    ids = pd.DataFrame(list(df['follower_id'].unique()), columns=['follower_id'])
    for i in tqdm(range(a, b)):
        ass = assing.rename({'author_id': 'influencer_id'}, axis=1)[['influencer_id', f'treatment_p{str(i)}']]
        ass['influencer_id'] = ass['influencer_id'].astype(str)
        dat = df.merge(ass, on='influencer_id', how='left')
        tie = 'weak_tie'
        weak = dat.groupby(['follower_id', f'treatment_p{str(i)}', tie]).count().reset_index()[['follower_id', f'treatment_p{str(i)}', tie, 'influencer_id']]
        weak = weak.pivot(index='follower_id', columns=[f'treatment_p{str(i)}', tie], values='influencer_id')
        weak = weak.reset_index()
        weak.loc[-1] = ['follower_id', f'n_influencers_followed_control_no_{tie}_p{str(i)}', f'n_influencers_followed_treatment_no_{tie}_p{str(i)}', f'n_influencers_followed_treatment_{tie}_p{str(i)}', f'n_influencers_followed_control_{tie}_p{str(i)}']  # adding a row
        weak.index = weak.index + 1  
        weak = weak.sort_index()  
        header_row = 0
        weak.columns = weak.iloc[header_row]
        weak = weak[1:].reset_index(drop=True)
        weak = weak.fillna(0)
        weak.reset_index(drop=True, inplace=True)
        weak[f"n_influencers_followed_control_p{str(i)}"] = weak.filter(like="control").sum(axis=1)
        weak[f"n_influencers_followed_treatment_p{str(i)}"] = weak.filter(like="treatment").sum(axis=1)
        weak['follower_id'] = weak['follower_id'].astype(str)
        influe_followed = df.groupby(['follower_id']).count().reset_index()[['follower_id', 'influencer_id']]
        tie = 'strong_tie'
        strong = dat.groupby(['follower_id', f'treatment_p{str(i)}', tie]).count().reset_index()[['follower_id', f'treatment_p{str(i)}', tie, 'influencer_id']]
        strong = strong.pivot(index='follower_id', columns=[f'treatment_p{str(i)}', tie], values='influencer_id')
        strong = strong.reset_index()
        strong.loc[-1] = ['follower_id', f"n_influencers_followed_control_no_{tie}_p{str(i)}", f'n_influencers_followed_treatment_no_{tie}_p{str(i)}', f'n_influencers_followed_control_{tie}_p{str(i)}', f'n_influencers_followed_treatment_{tie}_p{str(i)}']  # adding a row
        strong.index = strong.index + 1  # shifting index
        strong = strong.sort_index()     # sorting by index
        header_row = 0
        strong.columns = strong.iloc[header_row]
        strong = strong[1:].reset_index(drop=True)
        strong = strong.fillna(0)
        strong.reset_index(drop=True, inplace=True)
        batch = weak.merge(strong, on = 'follower_id', how='left')
        batch[f'n_influencers_followed_p_{str(i)}'] = influe_followed['influencer_id']
        ids = ids.merge(batch, on = 'follower_id', how='left')
    ids.to_parquet(f'../../data/04-analysis/{ba}/ties{e}.parquet')
    
    
```
