```python
import pandas as pd
import numpy as np
import random

from tweetple import TweetPle

from datetime import date, timedelta
from tqdm import tqdm
tqdm.pandas()
import time
import requests
import glob
import os

country = 'KE'
def get_path(country):
    base = f'../../data/01-characterize/followers/{country}/'
    path_save_c = base + '00-raw/collect/'
    path_save_i = base + '00-raw/integrate/'

    return path_save_c, path_save_i

path_save_c, path_save_i = get_path(country)

# Run R Script get_users_info() to get the ids, import them:
df = pd.read_parquet(f'../../data/01-characterize/influencers/{country}/confirmed_influencers_batch2.parquet')
ids = list(df['id'].astype(str))
len(ids)
# Total Number from Excel File for confirmation: SA 76 KE 76
```

```python
df_final = pd.DataFrame()
for name in ids:
    df = pd.read_parquet(f'{path_save_c}{name}.parquet')
    df_final = pd.concat([df_final, df]).reset_index(drop=True)
    
df_final.info()
```
